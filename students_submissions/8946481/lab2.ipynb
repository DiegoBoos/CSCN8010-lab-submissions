# **Irregularly gridded data**

### This figure is a modified version of **tripcolor**, with no axes and colorful

```python
import matplotlib.pyplot as plt
import numpy as np


# make data:
np.random.seed(1)
x = np.random.uniform(0, 1, 50)
y = np.random.uniform(0, 1, 50)
z = (1 - x/2 + x**5 + y**5) * np.exp(-x**10 - y**10)

# plot:
fig, ax = plt.subplots()

ax.plot(x, y, 'o', markersize=2, color='pink')
ax.tripcolor(x, y, z)

ax.set(xlim=(0, 1), ylim=(0, 1))

ax.axis('off')

plt.show()
```

# **Hexagonal binned plot**

### This figure is a **2D histogram plot**, with diferent sizes of hexagons

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

n = 100_000
x = np.random.standard_normal(n)
y = 2.0 + 3.0 * x + 4.0 * np.random.standard_normal(n)
xlim = x.min(), x.max()
ylim = y.min(), y.max()

fig, (ax0, ax1) = plt.subplots(ncols=2, sharey=True, figsize=(9, 4))

hb = ax0.hexbin(x, y, gridsize=10, cmap='inferno')
ax0.set(xlim=xlim, ylim=ylim)
ax0.set_title("Hexagon binning")
cb = fig.colorbar(hb, ax=ax0, label='counts')

hb = ax1.hexbin(x, y, gridsize=50, bins='log', cmap='inferno')
ax1.set(xlim=xlim, ylim=ylim)
ax1.set_title("With a log color scale")
cb = fig.colorbar(hb, ax=ax1, label='log10(N)')

plt.show()
```

# **Bivariate Gaussian**

### This figure is a **simulated data from a Bivariate Gaussiam**, with diferent sizes of hexagons and greater number of data points

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")

# Simulate data from a bivariate Gaussian
n = 50000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(1)
x, y = rng.multivariate_normal(mean, cov, n).T

# Draw a combo histogram and scatterplot with density contours
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=30, pthresh=.1, cmap="inferno")
sns.kdeplot(x=x, y=y, levels=10, color="pink", linewidths=1)
```

# **Map Overlay**

### This figure is a **Map application**, using the Dash framework and displays a map with points on it

```python
from dash import Dash, html
import holoviews as hv
from holoviews.plotting.plotly.dash import to_dash
from holoviews.element.tiles import CartoDark
from plotly.data import carshare

# Convert from lon/lat to web-mercator easting/northing coordinates
df = carshare()
df["easting"], df["northing"] = hv.Tiles.lon_lat_to_easting_northing(
    df["centroid_lon"], df["centroid_lat"]
)

points = hv.Points(df, ["easting", "northing"]).opts(color="white")
tiles = CartoDark()
overlay = tiles * points

app = Dash(__name__)
components = to_dash(app, [overlay])

app.layout = html.Div(
    components.children
)

if __name__ == '__main__':
    app.run(debug=True)
```
